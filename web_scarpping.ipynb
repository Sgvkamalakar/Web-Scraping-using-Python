```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
s=[]
links = soup.find_all('a')
for link in links:
  href = link.get('href')
  if href is not None and href.startswith('/'):
    s.append(href)
s=set(s)     
links=[]        
for i in s:
  if len(i)>40:
    print('https://www.amazon.in'+i)
    links.append('https://www.amazon.in'+i)
links=list(set(links)) 
links
```

```python
links=links[1:]
len(links)
```

```python
import csv
def to_csv(list1,list2,list3,list4,list5,filename):
  combined_column = zip(list1,list2,list3,list4,list5)
  with open(filename, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['URL', 'Product Name', 'Price (in ₹)', 'Prodcut ratings', 'Number of reviews'])  
    writer.writerows(combined_column)
  print(f"Data has been successfully written to {filename}")
```

```python
product_name,product_url,product_price,product_price,product_rating,product_reviews=[],[],[],[],[],[]
for url in links:
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  span_elements = soup.find_all('span', class_='a-size-base-plus a-color-base a-text-normal')
  product_name.extend([span.text.strip() for span in span_elements])
  print(url,len(product_name))
  elements = soup.find_all('a', class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal', target='_blank')
  for element in elements:
    if element.has_attr('href'):
      href = element['href']
      href='www.amazon.in'+href
      print(url,len(product_url))
      product_url.append(href)
  elements= soup.find_all('span', class_='a-price-whole')
  product_price.extend([span.text.strip() for span in elements])
  print(url,len(product_price))
  elements= soup.find_all('span', class_='a-icon-alt')
  product_rating.extend([span.text.strip() for span in elements])
  print(url,len(product_rating))
  elements= soup.find_all('span', class_='a-size-base s-underline-text')
  product_reviews.extend([span.text.strip() for span in elements])
  print(url,len(product_reviews))
       
to_csv(product_url,product_name,product_price,product_rating,product_reviews,'output.csv')      
```
